```python
# Import csv file
from pathlib import Path
import csv

menufile = Path(r'C:\Users\19802\Desktop\UNCC_Bootcamp\Homework\python-homework\PyRamen\resource_file\menu_data.csv')
salesfile = Path(r'C:\Users\19802\Desktop\UNCC_Bootcamp\Homework\python-homework\PyRamen\resource_file\sales_data.csv')

#Initialize variable as empty list
menu_data = {}
sales_data = []

# Open the menu csv file as an object
with open(menufile,'r') as menu_csv:
    
    #Create csv reader
    menu_reader = csv.reader(menu_csv)
    
    #To skip the header
    header = next(menu_reader)
              
    #@TODO: Iterate over each element of the list
    for row in menu_reader:
        menu_data[row[0]] = [row[1], row[2], row[3], row[4]]
        
# Open the sale file as an object
                             
with open(salesfile,'r') as sales_csv:
    
    #Create csv reader
    sales_reader = csv.reader(sales_csv)
    
    #To skip the header
    header = next(sales_reader)
    count = 0
              
    #@TODO: Iterate over each element of the list
    for row in sales_reader:
        sales_data.append([row[0], row[1], row[2], row[3], row[4]])
        
# Initialize report dictionary

report = {}

for sales in sales_data:
    
    quantity = int(sales[-2])
    sale_item = sales[-1]
    
    if report.get(sale_item)==None:
        report[sale_item] = {"01-count": 0, 
                             "02-revenue": 0, 
                             "03-cogs": 0, 
                             "04-profit": 0,
                            }
        
    menu = menu_data[sale_item]
    if menu!=None:
        price = float(menu[2])
        cost = float(menu[3])
        profit = quantity * (price - cost)
        report[sale_item]["01-count"] += quantity
        report[sale_item]["02-revenue"] += price * quantity
        report[sale_item]["03-cogs"] += cost * quantity
        report[sale_item]["04-profit"] += profit * quantity

        
# set the results in output file  

with open(Path("new_output.txt"), 'x') as outputfile:  
    for key, value in report.items():
        outputfile.write(str(key)+ ' ' + str(value) + '\n')


            






    
    
```
